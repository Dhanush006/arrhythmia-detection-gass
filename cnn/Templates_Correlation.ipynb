```python
import os
import cv2
import shutil
import numpy as np
from os import listdir
import matplotlib.pyplot as plt
import scipy.io
from scipy.interpolate import interp1d
```

```python
Dir = '/content/Signals_Resized_Today/Resized Signals/N78/'
Files=os.listdir(Dir)

names=[]

for y in Files:
    if y.endswith(".mat"):
        names.append(y)

final=[]

while (len(names)!=0):
    s1 = scipy.io.loadmat(Dir + names[0])
    print(len(names))
    indices=[]
    s1=s1["var"]
    for index, j in enumerate(names[1:]):
        s2 = scipy.io.loadmat(Dir + j)
        s2=s2["var"]
        r = np.corrcoef(s1.reshape(384), s2.reshape(384))[0][1]
        if (r>0.90):
            indices.append(index)
    for i in indices[::-1]:
        names.remove(names[i])
    final.append(names[0])
    names.remove(names[0])
```

```python
print("Number of templates for the class : ", len(final))
```

```python
!mkdir Temp
!mkdir Output
```

```python
Temp = '/content/Temp/'
Output = '/content/Output/'

for name in final:
    mat = scipy.io.loadmat(Dir + name)
    array = mat["var"][0]
    fig = plt.figure(figsize=(2, 2), dpi=80, frameon=False)
    x = np.linspace(1, len(array), len(array))
    plt.plot(x, array)
    plt.xticks([])
    plt.yticks([])
    fig.savefig(Temp + name[:-4] + ".png")
    plt.close()
    
    im_gray = cv2.imread(Temp + name[:-4] + ".png", cv2.IMREAD_GRAYSCALE)
    im_gray = cv2.resize(im_gray, (128, 128), interpolation = cv2.INTER_LANCZOS4)

    cv2.imwrite(Output + name[:-4] + ".png", im_gray)
```
